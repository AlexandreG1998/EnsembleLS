```python
def find_cointegrate (data = None, siglevel = 5, cut = 0.5):
    
    var_dependente = []
    var_independente = []
    sig_lvl = []
    split = int(len(data) * cut)
    
    treino = data[:split]
    for i in range (data.shape[1]):
        print('============================================')

        for k in range(data.shape[1]):
         
            a = treino.iloc[:,i:i+1]
            b = treino.iloc[:,k:k+1]
            try:

                model = RandomForestRegressor(max_depth= 30, random_state = 0).fit(a.values,b.values.ravel())
                prev  = model.predict(a.values)
                resid = pd.DataFrame(b.values.ravel()) - pd.DataFrame(prev)
                sig = adfuller(resid.values.ravel())[1] * 100

                var_dependente.append(a.columns[0])
                var_independente.append(b.columns[0])
                sig_lvl.append(sig)
            except:
                print('Erro')
            

            # Modelo Invertido
            try:
                model = RandomForestRegressor(max_depth= 30, random_state = 0).fit(b.values,a.values.ravel())
                prev  = model.predict(b.values)
                resid = pd.DataFrame(a.values.ravel()) - pd.DataFrame(prev)
                sig = adfuller(resid.values.ravel())[1] * 100

                var_dependente.append(b.columns[0])
                var_independente.append(a.columns[0])
                sig_lvl.append(sig)
            except:
                print('Erro')
            
                

        

    combinacoes = pd.concat((pd.DataFrame(var_dependente),
                            pd.DataFrame(var_independente),
                            pd.DataFrame(sig_lvl)), axis = 1)
    combinacoes.columns = ['Dependente','Independente','ADF']
    
    return(combinacoes)
```
